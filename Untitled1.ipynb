```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
pagi_count = soup.findAll('a', attrs={'class':'page'})
print(len(pagi_count)+1)
```

```python

i = 1
active_job_list = []
while i < len(pagi_count)+2:
    pagi_links = "http://www.jobsnepal.com/active-jobs/page-{}".format(i)
    req = requests.get(pagi_links)
    soup = BeautifulSoup(req.text,'lxml')
    data = soup.findAll("a",{"class":"job-item"})
    for item in data:
        link = item.get('href')
        active_job_list.append(link)
        print("Pagination {} link: {}".format(i,link))
    i +=1
   
```

```python
for job_link in active_job_list:
    
    req = requests.get(job_link)
    soup = BeautifulSoup(req.text,'lxml')
    data = soup.findAll('span',{'class':'apply-before-date'})
    general_data = soup.findAll('div',{'class':'detail-text'})
    for i in general_data:
#         print(i.text)
        job_detail = i.text
        if 'ios' in job_detail.lower():
            print(job_link)
        else:
            print("False")
        
        
    
```
