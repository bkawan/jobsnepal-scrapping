```python
import requests
from bs4 import BeautifulSoup
```

```python
link = "http://www.jobsnepal.com/active-jobs"
```

```python
r = requests.get(link)
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
pagi_count = soup.findAll('a', attrs={'class':'page'})
print(len(pagi_count)+1)
# pagi_count = [1,2,]
# print(len(pagi_count))
```

```python
i = 1
active_job_title= []
active_job_list = []
while i < len(pagi_count)+2:
    pagi_links = "http://www.jobsnepal.com/active-jobs/page-{}".format(i)
    req = requests.get(pagi_links)
    soup = BeautifulSoup(req.text,'lxml')
    data = soup.findAll("a",{"class":"job-item"})
    for item in data:
        active_job_title.append(item.string.strip())
        link = item.get("href")
        active_job_list.append(link)
        print('Pagination {} Job Title: {} link: {}'.format(i,item.string.strip(),link))
    i +=1
```
